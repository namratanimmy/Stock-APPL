# STOCK PRICE PREDICTION

### Importing dataset

```python
import pandas as pd
import sklearn
import numpy as np
dataset=pd.read_csv("HistoricalData_1623748350213.csv",parse_dates=["Date"],index_col="Date")
```

```python
dataset.head(10)
```

### Handling null values

```python
missing = dataset.isnull().sum()
missing
```

### striping

```python
#dataset['date']=pd.to_datetime(dataset['Date'],format="%m/%d/%Y")
```

```python
dataset['Open'].str.lstrip()
dataset['High'].str.lstrip()
dataset['Low'].str.lstrip()
dataset['Close/Last'].str.lstrip()
dataset.head()
```

### removing$

```python
dataset['Open']=dataset['Open'].str.replace('$',"")
dataset['Close/Last']=dataset['Close/Last'].str.replace('$',"")
dataset['High']=dataset['High'].str.replace('$',"")
dataset['Low']=dataset['Low'].str.replace('$',"")
dataset.head()
```

```python
dataset['Open'] = dataset['Open'].astype(float)
dataset['High'] = dataset['High'].astype(float)
dataset['Low'] = dataset['Low'].astype(float)
dataset['Close/Last'] = dataset['Close/Last'].astype(float)
dataset.head()
```
